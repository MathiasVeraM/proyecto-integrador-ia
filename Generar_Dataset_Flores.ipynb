```python
import pandas as pd
import numpy as np

# 1. Configurar semilla para reproducibilidad
np.random.seed(42)
n_muestras = 1000

# 2. Generar variables predictoras (Features del Dataset)
temperatura = np.random.uniform(12, 22, n_muestras)       # 12°C a 22°C
humedad_relativa = np.random.uniform(70, 98, n_muestras)  # 70% a 98%
radiacion_solar = np.random.uniform(80, 250, n_muestras)  # 80 a 250 W/m²

# Variables de ruido técnico (No correlacionadas)
velocidad_viento = np.random.uniform(0.5, 5.5, n_muestras)
presion_atmosferica = np.random.uniform(1008, 1014, n_muestras)
nivel_co2 = np.random.uniform(350, 450, n_muestras)

# 3. Construir el DataFrame
df = pd.DataFrame({
    'Temperatura_C': temperatura,
    'Humedad_Relativa_Porc': humedad_relativa,
    'Radiacion_Solar_W_m2': radiacion_solar,
    'Velocidad_Viento_kmh': velocidad_viento,
    'Presion_Atmosferica_hPa': presion_atmosferica,
    'Nivel_CO2_ppm': nivel_co2
})

# 4. Establecer el TARGET 1 (Clasificación binaria: Presencia del Hongo)
# Basado en la ley biológica real de la Botrytis cinerea (Humedad > 85% y Temp entre 15 y 18°C)
df['Presencia_Hongo'] = np.where(
    (df['Humedad_Relativa_Porc'] > 85) &
    (df['Temperatura_C'] >= 15) &
    (df['Temperatura_C'] <= 18),
    1, # Sí hay hongo
    0  # No hay hongo
)

# 5. Establecer el TARGET 2 (Regresión continua: Impacto / % Plantas Afectadas)
# Si Presencia_Hongo es 1, el impacto escala proporcionalmente a la humedad + ruido de la naturaleza.
# Si Presencia_Hongo es 0, el impacto es 0.0% (No hay afectación porque no hay hongo).
df['Porcentaje_Plantas_Afectadas'] = np.where(
    df['Presencia_Hongo'] == 1,
    (df['Humedad_Relativa_Porc'] - 85) * 6.5 + np.random.uniform(5, 18, n_muestras),
    0.0
)

# Asegurar que el porcentaje no supere el 100% por el ruido añadido
df['Porcentaje_Plantas_Afectadas'] = df['Porcentaje_Plantas_Afectadas'].clip(upper=100)

# Redondear valores para simular sensores reales
df = df.round({
    'Temperatura_C': 1,
    'Humedad_Relativa_Porc': 1,
    'Radiacion_Solar_W_m2': 0,
    'Velocidad_Viento_kmh': 2,
    'Presion_Atmosferica_hPa': 1,
    'Nivel_CO2_ppm': 0,
    'Presencia_Hongo': 0, # Queda como entero (0 o 1)
    'Porcentaje_Plantas_Afectadas': 1
})

# 6. Guardar el archivo definitivo para cargar en Orange
df.to_csv('dataset_flores.csv', index=False)
print("¡Dataset en cascada generado con éxito como 'dataset_cascada_botrytis.csv'!")
print("\nMuestra de los primeros 5 datos:")
print(df.head())
```
